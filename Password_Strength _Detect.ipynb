# Detecting the strength of the password.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### Reading dataset

```python
data=pd.read_csv('data.csv',error_bad_lines=False)
data.head()
```

#### shuffling data randomly for robustness

```python
import random
random.shuffle(password_tuple)
```

```python
x=[labels[0] for labels in password_tuple]#storing column1 in x
y=[labels[1] for labels in password_tuple]#storing column2 in y(target)
```

```python
x
```

```python
y
```

#### create a custom function to split input into characters of list

```python
def divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
divide_char('kino3434')
```

TF-TermFrequency
IDF-Inverse document frequency
It is basically multiplication of TF matrix into IDF matrix this give a vector representation of data that we have to pass to all ml algos depending on what use case we have. It assigns importance of each data.

#### import TF-IDF vectorizer to convert String data into numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector=TfidfVectorizer(tokenizer=divide_char)
```

#### apply TF-IDF vector on data

```python
X=vector.fit_transform(x)
```

```python
X.shape
```

```python
vector.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()#data
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vector.get_feature_names(),columns=['TF-IDF'])
df
```

```python
df.sort_values(by=['TF-IDF'],ascending=False)
```

#### split data into train & test
    train---> To learn the relationship within data, 
    test-->  To do predictions, and this testing data will be unseen to my model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

#### Apply Logistic on data as use-case is Classification

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

#### doing prediction for specific custom data

```python
dt=np.array(['%@123abcd'])
pred=vector.transform(dt)
clf.predict(pred)
```

#### doing prediction on X-Test data

```python
y_pred=clf.predict(X_test)
y_pred
```

#### check Accuracy of your model using confusion_matrix,accuracy_score

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
sns.heatmap(cm, annot=True, cmap='BuPu')
```

##### create report of your model

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
